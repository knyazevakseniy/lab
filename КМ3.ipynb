```python
import numpy as np
import CoolProp.CoolProp as ct
import matplotlib.pyplot as plt

t=np.linspace(0,381.15,100)
Fluid_name='CH4'

V=64
p=101325
d=ct.PropsSI('D','P',p,'T',t,Fluid_name)
m=d*V
t1=ct.PropsSI('T','P',p,'Q',1,Fluid_name)
d1=ct.PropsSI('D','P',p,'T',111.667,Fluid_name)
m1=d1*V
plt.plot(t,m,'blue')
plt.scatter(111.667,27030.78,color="black")
plt.grid()
plt.show()
print(m)
print("t1=",t1)
print('m1=',m1)
```
